# Matplotlib Object Oriented Method

```python
import matplotlib.pyplot as plt
```

```python
# Used to see plots on the jupyter notebook. otherwise u should use(plt.show()) on every plot.
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# Create an empty figure object.
fig = plt.figure()
# Add set of axes to figure
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left-padding, bottom-padding, width, height (range 0 to 1)
axes1.plot(x, y, 'b')
# Notice the use of set_ to begin methods
axes1.set_xlabel('X_label_axes')
axes1.set_ylabel('Y_label_axes')
axes1.set_title('Larger Plot')

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes2.plot(x, y, 'r')
axes2.set_xlabel('X_label_axes')
axes2.set_ylabel('Y_label_axes')
axes2.set_title('Inside Plot');
```

# subplots()

```python
# Use similar to plt.figure() except using this methods adds axes already.
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# Axes is an array of axes to plot on
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# seperates the overlapped figures from eachother.
plt.tight_layout

# Display the figure object    
fig
```

```python
fig, axes = plt.subplots(figsize=(12,3))# figsize used to specify the width and height of a figure

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig.savefig("filename.png") # stores the plot as a png file supports other extensions such as jpg, etc..
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.set_xlabel("x_axis")
ax.set_ylabel("y_axis")

ax.legend(loc=0) # loc specifies the location of putting the labels using 0 means best location found in pic.
```

# Setting colors, linewidths, linetypes

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

# Plot range

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
